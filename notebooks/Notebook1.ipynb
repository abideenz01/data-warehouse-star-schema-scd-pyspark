# DATA WAREHOUSING 

```sql
CREATE DATABASE orderDWH
```

### STAGING 

```sql
-- initial load 

CREATE OR REPLACE TABLE orderDWH.stg_sales
AS
SELECT * FROM sales_new.Orders


```

# TRANSFORMATION 

```sql
CREATE VIEW orderDWH.trans_sales
AS
SELECT * FROM orderDWH.stg_sales WHERE Quantity IS NOT NULL
```

```sql
SELECT * FROM orderDWH.trans_sales
```

# CORE LAYER 

### DIM_CUSTOMER 

```sql
CREATE OR REPLACE TABLE orderDWH.dim_Customers
(
    CustomerID INT,
    DimCustomerKey INT,
    CustomerName STRING,
    CustomerEmail STRING
);
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimCustomers AS
SELECT 
    T.CustomerID,
    T.CustomerName,
    T.CustomerEmail,
    ROW_NUMBER() OVER (ORDER BY T.CustomerID) AS DimCustomersKey
FROM
(
    SELECT DISTINCT
        CustomerID,
        CustomerName,
        CustomerEmail
    FROM orderDWH.trans_sales
) T;






```

```sql
INSERT INTO orderDWH.dim_Customers 
(
    CustomerID, 
    DimCustomerKey, 
    CustomerName, 
    CustomerEmail
)
SELECT 
    CustomerID, 
    DimCustomersKey, 
    CustomerName, 
    CustomerEmail
FROM orderDWH.view_DimCustomers;
```

```sql
SELECT * FROM orderDWH.dim_Customers
```

### DIM_PRODUCT

```sql
CREATE OR REPLACE TABLE orderDWH.dim_Products
(
    ProductID INT,
    DimProductsKey INT,
    ProductName STRING,
    ProductCategory STRING
);
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimProducts AS
SELECT 
    T.ProductID,
    T.ProductName,
    T.ProductCategory,
    ROW_NUMBER() OVER (ORDER BY T.ProductID) AS DimProductsKey
FROM
(
    SELECT DISTINCT
        ProductID,
        ProductName,
        ProductCategory
    FROM orderDWH.trans_sales
) T;
```

```sql
INSERT INTO orderDWH.dim_Products 
(
    ProductID, 
    DimProductsKey, 
    ProductName, 
    ProductCategory
)
SELECT 
    ProductID, 
    DimProductsKey, 
    ProductName, 
    ProductCategory
FROM orderDWH.view_DimProducts;
```

```sql
SELECT * FROM orderDWH.dim_Products
```

### DIM_REGION 

```sql
CREATE OR REPLACE TABLE orderDWH.dim_Region
(
    RegionID INT,
    DimRegionKey INT,
    RegionName STRING,
    Country STRING
);
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimRegion AS
SELECT 
    T.RegionID,
    T.RegionName,
    T.Country,
    ROW_NUMBER() OVER (ORDER BY T.RegionID) AS DimRegionKey
FROM
(
    SELECT DISTINCT
        RegionID,
        RegionName,
        Country
    FROM orderDWH.trans_sales
) T;




```

```sql
INSERT INTO orderDWH.dim_Region 
(
    RegionID, 
    DimRegionKey, 
    RegionName, 
    Country
)
SELECT 
    RegionID, 
    DimRegionKey, 
    RegionName, 
    Country
FROM orderDWH.view_DimRegion;
```

#### DIM_DATE

```sql
CREATE OR REPLACE TABLE orderDWH.dim_Date
(
    OrderDate Date,
    DimDateKey INT
);
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimDate AS
SELECT 
    OrderDate,
    ROW_NUMBER() OVER (ORDER BY OrderDate) AS DimDateKey
FROM
    (SELECT DISTINCT OrderDate FROM orderDWH.trans_sales) T;
```

```sql
INSERT INTO orderDWH.dim_Date (OrderDate, DimDateKey)
SELECT OrderDate, DimDateKey
FROM orderDWH.view_DimDate;
```

### FACT TABLE

```sql
CREATE  OR REPLACE TABLE orderDWH.fact_Sales
(


OrderID INT,
Quantity DECIMAL,
UnitPrice Decimal,
TotalAmount Decimal,
DimProductKey INT,
DimCustomersKey INT,
DimRegionKey INT,
DimDateKey INT
)
```

```sql
SELECT
F.OrderID,
F.Quantity,
F.UnitPrice,
F.TotalAmount,
DC.DimCustomerKey,
DP.DimProductsKey,
DR.DimRegionKey,
DD.DimDateKey

FROM orderDWH.trans_sales F
LEFT JOIN 
orderDWH.dim_Customers DC
ON F.CustomerID = DC.CustomerID
LEFT JOIN 
orderDWH.dim_Products DP
ON F.ProductID = DP.ProductID
LEFT JOIN
orderDWH.dim_Region DR
ON DR.Country = F.Country
LEFT JOIN
orderDWH.dim_Date DD
ON F.OrderDate = DD.OrderDate


```
